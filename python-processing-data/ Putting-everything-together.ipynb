```python
def wrapper_label_stock(index, data, tix):
    tix_data = search_by_ticker(data, tix)
    stock_tix = search_by_index(index, tix_data, 'Log Stock Return')
    index_tix = search_by_index(index, tix_data, 'Log Index Return')
    
    splitStock, splitIndex, cracks = split_by_cusum(stock_tix, index_tix, tix, False)
    labels = []
    for i in range(len(splitStock)):
        labels.append(tuple((cracks[i], int(one_tailed_two_sample_t_test(splitStock[i], \
            splitIndex[i], alpha=0.05)))))
    return labels
```

```python
def wrapper_label_full_stock(index, data, tix):
    start = 1
    tix_data = search_by_ticker(data, tix)
    stock_tix = search_by_index(index, tix_data, 'Log Stock Return').astype('float32')
    index_tix = search_by_index(index, tix_data, 'Log Index Return').astype('float32')
    
    splitStock, splitIndex, cracks = split_by_cusum(stock_tix, index_tix, tix, False)
    labels = []
    for i in range(len(cracks)-1):
        start = cracks[i]
        end   = cracks[i+1]
        while (end-start) >= 10:
            tmp = tuple((start, int(one_tailed_two_sample_t_test(stock_tix[start:end], \
                index_tix[start:end], alpha=0.05))))
            labels.append(tmp)
            start += 1
    return labels
```
