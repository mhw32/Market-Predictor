## CUSUM Algorithm applied to the Financial Setting

Most of this is based on the link:<br> http://nbviewer.ipython.org/github/demotu/BMC/blob/master/notebooks/DetectCUSUM.ipynb
Edits have been made to the code, but the foundation is borrowed.

<b>The goal here is to detect changes with Cumulative Sum</b>

Tuning steps:
- Start with a very large threshold.
- Choose drift to one half of the expected change, or adjust drift such that g = 0 more than 50% of the time.
- Then set the threshold so the required number of false alarms (this can be done automatically) or delay for detection is obtained.
- If faster detection is sought, try to decrease drift.
- If fewer false alarms are wanted, try to increase drift.
- If there is a subset of the change times that does not make sense, try to increase drift.

```python
# This function just recurses through possible options in small steps and stores the good ones.
def searchParams(data, step, maxim):
    ending, show, ax = True, False, None
    # temporary variables
    init_thres = maxim
    init_drift = 0.0
    # storage variables
    possible_params = []
    
    while init_thres >= 0:
        while init_drift <= maxim:
            ta, tai, taf, amp = detect_cusum(stock_AA, \
                threshold, drift, ending, show, ax)
            if len(ta) > 0:
                possible_params.append(tup(init_thres, init_drift))
            
            init_drift = init_drift + step      
        init_thres = init_thres - step
        # Set the other thing back to normal
        init_drift = 0.0

    return possible_params
```

```python
def split_by_cusum(data, index, ticker, plotting=False):
    # Pick out the stock values through ticker
    tix = search_by_ticker(data, ticker)
    stock_tix = search_by_index(index, tix, 'Log Stock Return').astype('float32')
    # Apply the CUSUM
    threshold, drift, ending, show, ax = 0.10, 0.05, True, plotting, None
    ta, tai, taf, amp = detect_cusum(stock_tix, threshold, drift, ending, show, ax)
    # Append and Prepend the right values
    ta = list(ta)
    if (ta[-1] != len(stock_tix)):
        ta.append(len(stock_tix))
    ta = [0] + ta
    # Do the splitting
    split_by_cusum = np.array([stock_tix[i:j] for i,j in zip(ta[0:len(ta)-1], ta[1:])])
    return split_by_cusum
```

```python
def search_best_params(stock_data, plotting=False):
    combos = []
    # Try all the i, j combos for (threshold, drift)
    for i in np.linspace(0, 0.15, 10):
        for j in np.linspace(0, 0.15, 10):
            ta, tai, taf, amp = detect_cusum(stock_data, i, j, True, plotting, None)
            seg = [ta[k+1]-ta[k] for k in range(len(ta)-1)]
            # Let's try to pick out the ones with definitely more than 10
            seg = np.array(seg)
            # Make sure that there are no super small differenes, and none of them are 0
            if np.sum(seg < 20) > 0 or len(seg) == 0:
                continue
            # print('threshold: %f & drift: %f' % (i, j))
            combos.append((i, j, ta, seg))
    if len(combos) > 0:
        # Sort the combos by length and return values
        combos.sort(key = lambda s: len(s[3]))
        chosen_threshold, chosen_drift, chosen_cuts, _ = combos[::-1][0]
        return chosen_threshold, chosen_drift, chosen_cuts
    else:
        print('Error. Found nothing to fit the description provided.')
        return 0, 0, 0
```

```python
def split_by_cusum(data, index, ticker, plotting=False):
    #print('Starting CUSUM Evaluation')
    #print('-------------------------')
    # Pick out the stock values through ticker
    tix = search_by_ticker(data, ticker)
    stock_tix = search_by_index(index, tix, 'Log Stock Return').astype('float32')
    market_tix = search_by_index(index, tix, 'Log Index Return').astype('float32')
    # Apply the CUSUM
    #print ('Finding Optimal Parameter...')
    threshold, drift, ta = search_best_params(stock_tix)
    if type(ta) == int:
        return 0
    #print('Set threshold: %f & drift: %f...' % (threshold, drift))
    #print('Found %d divisions...' % len(ta))
    # Append and Prepend the right values
    ta = list(ta)
    if (ta[-1] != len(stock_tix)):
        if (len(stock_tix) - ta[-1] > 20):
            ta.append(len(stock_tix))
        else:
            ta[len(ta)-1] = len(stock_tix)
    if ta[0] > 20:
        ta = [0] + ta
    else:
        ta[0] = 0
    # Do the splitting
    split_by_cusum = np.array([stock_tix[i:j] for i,j in zip(ta[0:len(ta)-1], ta[1:])])
    split_by_market = np.array([market_tix[i:j] for i,j in zip(ta[0:len(ta)-1], ta[1:])])
    #print ('Complete: time horizons calculated and split successfully...')
    return split_by_cusum, split_by_market, ta
```

```python
# The goal now is to select 50, and only pick the ones out of there that are > 20 
# and consolidate the rest!
def cusum_consolidate(complete):
    joined = []
    indices, shifts = complete[2], complete[3]
    base = indices[0]
    for i in range(len(indices)-1):
        if indices[i] - base > 20:
            joined.append(indices[i])
            base = indices[i]
    joined = np.array(joined)
    return joined
```

```python
def search_best_params(stock_data, plotting=False):
    combos = []
    # Try all the i, j combos for (threshold, drift)
    for i in np.linspace(0, 0.25, 20):
        for j in np.linspace(0, 0.25, 20):
            ta, tai, taf, amp = detect_cusum(stock_data, i, j, True, plotting, None)
            seg = [ta[k+1]-ta[k] for k in range(len(ta)-1)]
            # Let's try to pick out the ones with definitely more than 10
            seg = np.array(seg)
            # Make sure that there are no super small differenes, and none of them are 0
            if len(seg) <= 40 or len(seg) >= 50:
                continue
            # print('threshold: %f & drift: %f' % (i, j))
            combos.append((i, j, ta, seg))
    if len(combos) > 0:
        # Sort the combos by length and return values
        combos.sort(key = lambda s: sum(s[3]))
        chosen_threshold, chosen_drift, chosen_cuts, _ = combos[::-1][0]
        chosen_cuts = cusum_consolidate(combos[::-1][0])
        return chosen_threshold, chosen_drift, chosen_cuts
    else:
        print('Error. Found nothing to fit the description provided.')
        return 0, 0, 0
```

#### We should try to write a CUSUM threshold labelling function

```python
# This is intended to be a function to categorize the data into 2/3 labels
# Small, Medium, Large intervals!!!
CUSUM_LABEL_LOW = 30
CUSUM_LABEL_HIGH = 60
def cusum_threshold(complete):
    small  = [] # let's say < 15
    medium = [] # 15 - 30
    large  = [] # > 40
    
    
    indices, shifts = complete[2], complete[3]
    base = indices[0]
    for i in range(len(indices)-1):
        step = indices[i] - base
        if  step <= CUSUM_LABEL_LOW:
            small.append(indices[i])
        elif step > CUSUM_LABEL_LOW and step <= CUSUM_LABEL_HIGH:
            medium.append(indices[i])
        else:
            large.append(indices[i])
    small = np.array(small)
    medium = np.array(medium)
    large = np.array(large)
    return small, medium, large
```
